```python
## Test with different SMA
def moving_average_crossover_strategy(data, short_window, long_window, atr_multiplier, rsi_overbought, rsi_oversold, fund, transaction_cost=10):
    buy_signal = []
    sell_signal = []
    open_position = []
    funds = [float(fund)] * len(data)
    last_funds = float(fund)
    
    buy_signal_bh = []
    sell_signal_bh = []
    open_position_bh = [float(fund) / data['Adj Close'].iloc[0]] * len(data)
    funds_bh = [float(fund)] * len(data)
    
    flag = 0  # flag = 0 means sell_flag and flag = 1 means buy_flag
    signals = pd.DataFrame(index=data.index)
    signals['price'] = data['Adj Close']
    signals['short_mavg'] = data['Adj Close'].rolling(window=short_window, min_periods=1, center=False).mean()
    signals['long_mavg'] = data['Adj Close'].rolling(window=long_window, min_periods=1, center=False).mean()
    signals['atr'] = ta.volatility.average_true_range(data['High'], data['Low'], data['Close'], window=14)
    signals['rsi'] = ta.momentum.RSIIndicator(data['Close'], window=14).rsi()
    
    flag = 0
    for i in range(len(data)):
        if (signals['short_mavg'][i] > signals['long_mavg'][i]) & (signals['rsi'][i] < rsi_overbought):
            if flag == 0:
                flag = 1
                buy_signal.append(signals['price'][i])
                last_pos = last_funds / signals['price'][i]
                funds[i] = last_funds - transaction_cost
                open_position.append(last_pos)     # buy_quantity with 1 Lac Capital
                sell_signal.append(np.NaN)
            else:
                buy_signal.append(np.NaN)
                last_funds = signals['price'][i] * last_pos
                funds[i] = last_funds
                open_position.append(last_pos)
                sell_signal.append(np.NaN)
        elif (signals['short_mavg'][i] > signals['long_mavg'][i]) & (signals['rsi'][i] < rsi_overbought):
            if flag == 1:
                flag = 0
                buy_signal.append(np.NaN)
                last_funds = last_pos * signals['price'][i] - transaction_cost
                funds[i] = last_funds
                open_position.append(0)
                sell_signal.append(signals['price'][i])
            else:
                buy_signal.append(np.NaN)
                funds[i] = last_funds
                open_position.append(0)
                sell_signal.append(np.NaN)
        else:
            buy_signal.append(np.NaN)
            open_position.append(0)
            sell_signal.append(np.NaN)
            funds[i] = last_funds
    
    # Calculate stop-loss levels based on ATR
    signals['stop_loss'] = signals['price'] - atr_multiplier * signals['atr']
    
    for i in range(1, len(data)):
        funds_bh[i] = funds_bh[i - 1] * (data['Adj Close'][i] / data['Adj Close'][i - 1])

    # At the end of the time period, simulate selling the entire remaining position
    funds_bh[-1] = data['Adj Close'].iloc[-1] * open_position_bh[-1]
    
    return signals, buy_signal, sell_signal, open_position, funds, flag, funds_bh[-1]
```

```python
def get_best_sma(data_to_download):
    best_sma_each_stock = []
    for key, values in data_to_download.items():
        for value in values:
            data = pd.read_csv(f'Data/{key} - {value}.csv')
            if len(data) == 0:
                continue
            short_windows = [5, 7, 9, 10, 20, 21, 30, 40, 50, 100]
            long_windows = [7, 9, 10, 20, 21, 30, 40, 50, 100, 200]
            atr_multiplier = 2
            rsi_overbought = 70
            rsi_oversold = 30
            initial_funds = 100000
            best_fund = 0
            best_short=0
            best_long=0
            for short_window in short_windows:
                for long_window in long_windows:
                    if long_window <= short_window:
                        continue
                    signals, buy_signal, sell_signal, open_position, funds, flag, funds_bh = moving_average_crossover_strategy(data, short_window, long_window, atr_multiplier, rsi_overbought, rsi_oversold, initial_funds)
                    if funds[-1] > best_fund:
                        best_fund = max(best_fund, funds[-1])
                        best_short = short_window
                        best_long = long_window

            best_sma_each_stock.append([key, value, best_short, best_long, best_fund, ((best_fund-funds[0])/funds[0] * 100), funds_bh])
    columns = ['industry', 'stock', 'Short_window', 'Long_window', 'Final_funds', 'Percentage_change', 'buy_and_hold']
    total = pd.DataFrame(best_sma_each_stock, columns=columns)
    return total
```

```python
def get_mode(series):
    return series.value_counts().idxmax()
```

```python
def backtest_result(total):
    test = total.groupby('industry')[['Short_window', 'Long_window']].apply(get_mode).reset_index()
    get_data_all_test(data_to_download, start_test, end_test)
    best_sma_ind = []
    for key, values in data_to_download.items():
        for value in values:
            data = pd.read_csv(f'Test/{key} - {value}.csv')
            (short_window, long_window) = test[test['industry'] == f'{key}'][0].values[0]
            if len(data) == 0:
                continue
            atr_multiplier = 2
            rsi_overbought = 70
            rsi_oversold = 30
            funds = 100000
            signals, buy_signal, sell_signal, open_position, funds, flag, funds_bh = moving_average_crossover_strategy(data, short_window, long_window, atr_multiplier, rsi_overbought, rsi_oversold, funds)
            best_sma_ind.append([key, value, short_window, long_window, funds[-1], ((funds[-1-funds[0])/funds[0] * 100), funds_bh])
    cols = ['industry', 'stock', 'Short_window', 'Long_window', 'Final_funds', 'Percentage_change', 'buy_and_hold']
    result = pd.DataFrame(best_sma_ind, columns=cols)
    return result
```

```python
def display_data(results):
    plt.figure(figsize = (15, 8))
    plt.plot(results['Final_funds'], label = 'Final Funds SMA', linewidth=1)
    plt.plot(results['buy_and_hold'], label = 'Final Funds Buy and Hold', linewidth=1)
    
    avg_sma = results['Final_funds'].mean()
    avg_bh = results['buy_and_hold'].mean()
    plt.axhline(y=avg_sma, color='blue', linestyle='--', label=f'Average SMA: {avg_sma:.2f}')
    plt.axhline(y=avg_bh, color='green', linestyle='--', label=f'Average Buy and Hold: {avg_bh:.2f}')
    
    
    plt.title('Comparison between SMA and Buy and Hold strategies')
    plt.xlabel('Stocks')
    plt.ylabel('Fund Amount')
    plt.legend(loc='upper left')
    plt.show()
```
